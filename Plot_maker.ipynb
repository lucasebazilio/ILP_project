# α Tuning

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file into a DataFrame
file_path = 'average_profits_n5000_m20.csv'  # Update with the correct file path
data = pd.read_csv(file_path)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data['Alpha'], data['Average Profit'], marker='o', linestyle='-')
plt.title('Average Profit as a Function of α for n = 5000 and m = 20')
plt.xlabel('Alpha')
plt.ylabel('Average Profit')
plt.grid(True)
plt.show()

# Find the alpha corresponding to the smallest average profit
max_alpha = data.loc[data['Average Profit'].idxmax()]['Alpha']

print(f"The value of alpha with the largest average profit is: {max_alpha}")
```

## Comparative Analysis


### Solution Quality

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths for each CSV file
file_paths = [
    'comparative_Greedy.csv',
    'comparative_Grasp_m20_alpha02.csv',
    'comparative_LocalSearch.csv',
    'comparative_cplex.csv'
]

# Initialize an empty list to store DataFrames
dfs = []

# Iterate over all file paths
for file_path in file_paths:
    # Read CSV file into a DataFrame
    df = pd.read_csv(file_path)
    dfs.append((file_path, df))  # Store file path along with DataFrame

# Plotting
plt.figure(figsize=(10, 6))

# Iterate over the list of DataFrames and plot Profit vs nOrders
for file_path, df in dfs:
    if file_path == 'comparative_Greedy.csv':
      my_label = 'Greedy'
    elif file_path == 'comparative_LocalSearch.csv':
      my_label = 'Local Search'
    elif file_path == 'comparative_Grasp_m20_alpha02.csv':
      my_label = 'GRASP'
    else:
      my_label = 'CPLEX'
    plt.plot(df['nOrders'], df['Profit'], marker='o', linestyle='-', label=my_label)

plt.title('Comparative Results (GRASP with m = 20 and α = 0.2)')
plt.xlabel('nOrders')
plt.ylabel('Profit')
plt.grid(True)
plt.legend()
plt.show()
```

### Computation Time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths for each CSV file
file_paths = [
    'comparative_Greedy.csv',
    'comparative_Grasp_m20_alpha02.csv',
    'comparative_LocalSearch.csv',
    'comparative_cplex.csv'
]

# Initialize an empty list to store DataFrames
dfs = []

# Iterate over all file paths
for file_path in file_paths:
    # Read CSV file into a DataFrame
    df = pd.read_csv(file_path)
    dfs.append((file_path, df))  # Store file path along with DataFrame

# Plotting
plt.figure(figsize=(10, 6))

# Iterate over the list of DataFrames and plot Profit vs nOrders
for file_path, df in dfs:
    if file_path == 'comparative_Greedy.csv':
      my_label = 'Greedy'
    elif file_path == 'comparative_LocalSearch.csv':
      my_label = 'Local Search'
    elif file_path == 'comparative_Grasp_m20_alpha02.csv':
      my_label = 'GRASP'
    else:
      my_label = 'CPLEX'
    plt.plot(df['nOrders'], df['Time'], marker='o', linestyle='-', label=my_label)

plt.yscale('log')  # Set y-axis to logarithmic scale
plt.title('Computation Time Comparison (GRASP with m = 20 and α = 0.2)')
plt.xlabel('nOrders')
plt.ylabel('Time (log scale)')
plt.grid(True)
plt.legend()
plt.show()
```
